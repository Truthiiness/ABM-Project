```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('traits.csv')
df.describe()
```

```python
is_fb = df[df['Organization']=='FancyBear'].index
df.drop(is_fb, inplace=True)
is_cb = df[df['Organization']=='CozyBear'].index
df.drop(is_cb, inplace=True)
is_veb = df[df['Organization']=='VenomousBear'].index
df.drop(is_veb, inplace=True)
is_vob = df[df['Organization']=='VoodooBear'].index
df.drop(is_vob, inplace=True)
is_bb = df[df['Organization']=='BerserkBear'].index
df.drop(is_bb, inplace=True)
is_pb = df[df['Organization']=='PrimitiveBear'].index
df.drop(is_pb, inplace=True)
df.describe()
```

```python
agent_data={}
for rownumber, row in df.iterrows():
    if row['AgentID'] in agent_data.keys():
        agent_data[row['AgentID']]['Spear Phishing'].append(row['Spear Phishing'])
    else:
        agent_data[row['AgentID']] = {'Spear Phishing': [row['Spear Phishing']]}

for rownumber, row in df.iterrows():
    if 'Zero Day' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Zero Day'].append(row['Zero Day'])
    else: 
        agent_data[row['AgentID']]['Zero Day'] = [row['Zero Day']]

for rownumber, row in df.iterrows():
    if 'Tool Sophistication' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Tool Sophistication'].append(row['Tool Sophistication'])
    else: 
        agent_data[row['AgentID']]['Tool Sophistication'] = [row['Tool Sophistication']]

for rownumber, row in df.iterrows():
    if 'Attribution Obfuscation' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Attribution Obfuscation'].append(row['Attribution Obfuscation'])
    else: 
        agent_data[row['AgentID']]['Attribution Obfuscation'] = [row['Attribution Obfuscation']]
        
for rownumber, row in df.iterrows():
    if 'Stealth' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Stealth'].append(row['Stealth'])
    else: 
        agent_data[row['AgentID']]['Stealth'] = [row['Stealth']]

for rownumber, row in df.iterrows():
    if 'Information Weaponization' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Information Weaponization'].append(row['Information Weaponization'])
    else: 
        agent_data[row['AgentID']]['Information Weaponization'] = [row['Information Weaponization']]

for rownumber, row in df.iterrows():
    if 'DDoS' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['DDoS'].append(row['DDoS'])
    else: 
        agent_data[row['AgentID']]['DDoS'] = [row['DDoS']]

for rownumber, row in df.iterrows():
    if 'Data Destruction' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Data Destruction'].append(row['Data Destruction'])
    else: 
        agent_data[row['AgentID']]['Data Destruction'] = [row['Data Destruction']]
        
for rownumber, row in df.iterrows():
    if 'Critical Infrastructure Disruption' in agent_data[row['AgentID']].keys():
        agent_data[row['AgentID']]['Critical Infrastructure Disruption'].append(row['Critical Infrastructure Disruption'])
    else: 
        agent_data[row['AgentID']]['Critical Infrastructure Disruption'] = [row['Critical Infrastructure Disruption']]
```

```python
def plot_average(agent_data, attribute):
    attribute_average = []
    for x in range(120):
        step = []
        agent_interested = set()
        for k, v in agent_data.items():
            if v[attribute][x] == 0:
                pass
            else:
                step.append(v[attribute][x])
                agent_interested.add(k)
        avg = sum(step)/len(agent_interested)
        attribute_average.append(avg)
    return attribute_average
```

```python
listoflists = {}
attributes = list(agent_data[2].keys())
for a in attributes:
    listoflists[a] = plot_average(agent_data, a)
```

```python
Step = list(range(1,121))
fig, ax = plt.subplots(figsize=(20, 15))
for k, v in listoflists.items():
    ax.scatter(Step, v, label=k)
ax.legend()
ax.set_ylabel('Traits')
ax.set_xlabel('Step')
plt.title('Iranian Agent Average')
```

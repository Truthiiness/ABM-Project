```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('traits.csv')
df.describe()
```

```python
ax1 = df.plot.scatter(x='Step', y='Spear Phishing', c='blue', s=.5, label='Phishing')
ax2 = df.plot.scatter(x='Step', y='Zero Day', c='darkblue', s=.5, label='Zero Day', ax=ax1)
ax3 = df.plot.scatter(x='Step', y='Tool Sophistication', c='darkgreen', s=.5, label='Tools', ax=ax1)
ax4 = df.plot.scatter(x='Step', y='Attribution Obfuscation', c='grey', s=.5, label='Obfuscation', ax=ax1)
ax5 = df.plot.scatter(x='Step', y='Stealth', s=.5, c='black', label='Stealth', ax=ax1)
ax6 = df.plot.scatter(x='Step', y='Information Weaponization', s=.5, c='red', label='Info Weap', ax=ax1)
ax7 = df.plot.scatter(x='Step', y='DDoS', s=.5, c='darkred', label='DDoS', ax=ax1)
ax8 = df.plot.scatter(x='Step', y='Data Destruction', s=.5, c='violet', label='Data Destruct', ax=ax1)
ax9 = df.plot.scatter(x='Step', y='Critical Infrastructure Disruption', c='purple', s=.5, label='Infrastructre', ax=ax1)
```
